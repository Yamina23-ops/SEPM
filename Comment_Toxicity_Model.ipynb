```python
from tensorflow.keras.layers import TextVectorization
```

```python
X = df['comment_text']
Y = df[df.columns[2:]].values
```

```python
df.columns
```

```python
df[df.columns[2:]]
```

```python
MAX_FEATURES = 200000
```

```python
vectorizer = TextVectorization(max_tokens = MAX_FEATURES,
                               output_sequence_length = 1800,
                               output_mode = 'int')
```

```python
vectorizer.adapt(X.values)
vectorized_text = vectorizer(X.values)
```

```python
vectorizer('Hello world, life is great')[:5]
```

```python
vectorizer.get_vocabulary()
```

```python
vectorized_text = vectorizer(X.values)
```

```python
len(X)
```

```python
vectorized_text
```

```python
# MCSHBAP - map, cache, shuffle, batch, prefetch from_tensor_slices, list_file
dataset = tf.data.Dataset.from_tensor_slices((vectorized_text, Y))
dataset = dataset.cache()
dataset = dataset.shuffle(160000)
dataset = dataset.batch(16)
dataset = dataset.prefetch(8) # helps bottlenecks
```

```python
batch_X, batch_Y = dataset.as_numpy_iterator().next()
```

```python
train = dataset.take(int(len(dataset)*.7))
val = dataset.skip(int(len(dataset)*.7)).take(int(len(dataset)*.2))
test = dataset.skip(int(len(dataset)*.9)).take(int(len(dataset)*.1))
```

```python
len(dataset)*.7
```

```python
import tensorflow as tf

train.as_numpy_iterator().next()
```

**CREATE SEQUENTIAL MODEL**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Bidirectional, Dense, Embedding
```

```python
Bidirectional??
```

```python
Y[0]
```

```python
model = Sequential()
#create embedding layer
model.add(Embedding(MAX_FEATURES+1,32))
# bidirectional layer
model.add(Bidirectional(LSTM(32, activation='tanh')))
# feature extraction fully connected layers
model.add(Dense(128, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
# final layer
model.add(Dense(6, activation='sigmoid'))
```

```python
model.compile(loss = 'BinaryCrossentropy', optimizer = 'Adam')
```

```python
model.summary()
```

```python
history = model.fit(train, epochs=1, validation_data=val)
```

```python
history.history
```

**EVALUATE MODEL**

```python
from tensorflow.keras.metrics import Precision, Recall, CategoricalAccuracy
```

```python
pre = Precision()
re = Recall()
acc = CategoricalAccuracy()
```

```python
for batch in test.as_numpy_iterator():
    # unpack the batch
    X_true, y_true = batch

    # make a prediction
    yhat = model.predict(X_true)

    y_true = y_true.flatten()
    yhat = yhat.flatten()

    pre.update_state(y_true, yhat)
    re.update_state(y_true, yhat)
    acc.update_state(y_true, yhat)
```

```python
print(f'Precision: {pre.result().numpy()}, Recall: {re.result().numpy()}, Accuracy: {acc.result().numpy()}')
```

```python
import gradio as gr

def score_comment(comment):
    vectorized_comment = vectorizer([comment])
    results = model.predict(vectorized_comment)

    output = ""
    for idx, col in enumerate(df.columns[2:]):
        label = "рҹ”ҙ Toxic" if results[0][idx] > 0.5 else "рҹҹў Safe"
        output += f"{col.upper()}: {label}\n"

    return output


with gr.Blocks(theme=gr.themes.Soft()) as app:

    gr.Markdown(
        """
        # Troll Detection System
        Enter a comment below to check if it is toxic or safe.
        """
    )

    with gr.Row():
        comment_input = gr.Textbox(
            lines=4,
            placeholder="Type your comment here...",
            label="рҹ’¬ Input Comment"
        )

    with gr.Row():
        output_box = gr.Textbox(
            label="рҹ“Ҡ Prediction Result",
            lines=10,        # рҹ‘Ҳ BIG OUTPUT BOX
            interactive=False
        )

    with gr.Row():
        submit_btn = gr.Button(" Analyze", variant="primary")
        clear_btn = gr.Button(" Clear")

    submit_btn.click(fn=score_comment, inputs=comment_input, outputs=output_box)
    clear_btn.click(lambda: ("", ""), None, [comment_input, output_box])

app.launch()
```
